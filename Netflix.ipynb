#### What is Netflix?
Netflix is an American subscription video on-demand over-the-top streaming service. The service primarily distributes original and acquired films and television shows from various genres, and it is available internationally in multiple languages.

Launched in 2007, nearly a decade after Netflix, Inc. began its pioneering DVD-by-mail movie rental service, Netflix is the most-subscribed video on demand streaming media service, with 282.7 million paid memberships in more than 190 countries as of 2024. By 2022, "Netflix Original" productions accounted for half of its library in the United States and the namesake company had ventured into other categories, such as video game publishing of mobile games through its flagship service. As of 2023, Netflix is the 23rd most-visited website in the world, with 23.66% of its traffic coming from the United States, followed by the United Kingdom at 5.84%, and Brazil at 5.64%.?

## Visual Analysis

#### 1) Univariate Analysis

```python
# Convert Duration to numerical values for movies
df["Minutes"] = df["duration"].apply(lambda x: int(x.split()[0]) if isinstance(x, str) and 'min' in x else None)
```

```python
sns.set(style="whitegrid")
```

```python
# Univariate Analysis
#For Continous Variable Release year
# Distribution of Release Year
plt.figure(figsize=(10, 5))
sns.histplot(df["release_year"], kde=True, color="skyblue", bins=30)
plt.title("Distribution of Release Year")
plt.xlabel("Release Year")
plt.ylabel("Frequency")
plt.show()
```

 ### ðŸ” Insights
 Distribution of Release Year:
 * The plot shows the distribution of movie release years, which is skewed towards the more recent years.
 * There is a clear peak in the number of movies released in the years around 2020. This suggests a significant increase in movie production in recent times.
 * The number of movies released gradually increased from the 1940s to the 2020s, with a steeper rise observed in the later years. This indicates a trend of increased movie production over time.
 * The number of movies released in the earlier years (1940s to 1970s) is significantly lower compared to the recent years, indicating a smaller movie production in those periods.

```python
# Distribution of Duration (Movies Only)
#For Continous Variable Duriation 
plt.figure(figsize=(10, 5))
sns.histplot(df[df['type'] == 'Movie']['Minutes'].dropna(), kde=True, color="purple", bins=30)
plt.title("Distribution of Movie Duration (Minutes)")
plt.xlabel("Duration (Minutes)")
plt.ylabel("Frequency")
plt.show()
```

 ### ðŸ” Insights
 Distribution of Movie Duration:
* The plot shows that most movies have a duration between 60 and 150 minutes, but there are a few movies that are much longer. This means the distribution is skewed to the right, with a long tail. This is the usual length for movies, probably because it helps keep viewers interested.
* some movies are significantly longer than the majority. These are the outliers in the plot.

```python
#Categorical variables like Type 
# Count of Movies vs TV Shows
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="type")
plt.title("Count of Movies vs TV Shows")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
```

 ### ðŸ” Insights
 Count of Movies vs. TV Shows:
 * The plot shows that Netflix has more movies than TV shows. There are roughly 6000 movies compared to around 2500 TV shows.
 * This could imply that movies are a more popular or common content type on Netflix compared to TV shows.
 * Netflix might prioritize acquiring more movies over TV shows, or this may be a reflection of the overall movie and TV show landscape.

```python
#Categorical variables like Listed_in(Genre)
# Top 10 Genres
plt.figure(figsize=(12, 6))
top_genres = df["listed_in"].value_counts().head(10)
sns.barplot(y=top_genres.index, x=top_genres.values, palette="magma")
plt.title("Top 10 Genres on Netflix")
plt.xlabel("Count")
plt.ylabel("Genre")
plt.show()
```

 ### ðŸ” Insights
Top 10 Genres:
* Dramas and international Movies are the most popular genres, followed closely by documentaries.
* Stand-up comedy is a popular genre, likely reflecting the growing demand for comedic content.
* Comedies, Dramas, International Movies and Dramas, Independent Movies, and International Movies indicate that movies with international flavor are a popular choice.
* Kids'TV is a popular genre, suggesting a large number of families use Netflix for their children's entertainment.
* Children & Family Movies and Children & Family Movies, Comedies suggest that Netflix offers a wide variety of family-friendly options.
* Documentaries, International Movies and Dramas, International Movies, and Romantic Movies suggest that viewers are interested in diverse stories and genres.

```python
#Categorical variables like Country
#Top 10 Countries by Content Production
plt.figure(figsize=(12, 6))
top_countries = df["country"].value_counts().head(10)
sns.barplot(y=top_countries.index, x=top_countries.values, palette="cool")
plt.title("Top 10 Countries by Content Production")
plt.xlabel("Count")
plt.ylabel("Country")
plt.show()
```

 ### ðŸ” Insights

 * The United States makes the most content for Netflix, followed by India, the UK, and Canada.
 * This shows that Netflix focuses a lot on content from these countries because they have large English-speaking audiences and strong film industries.

#### 2) Bivariate Analysis

```python
#Boxplot of Release Year by Type
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x="type", y="release_year", palette="Set2")
plt.title("Release Year by Type (Movies vs TV Shows)")
plt.xlabel("Type")
plt.ylabel("Release Year")
plt.show()
```

 ### ðŸ” Insights
 Boxplot of Release Year by Type:
 * TV Shows have a slightly more recent release year distribution than Movies.
 * This may suggest that Netflix's investment in new content is more geared towards the production of TV Shows instead of movies.

```python
#Boxplot of Movie Duration by Type
plt.figure(figsize=(10, 5))
sns.boxplot(data=df[df["type"] == "Movie"], x="type", y="Minutes", palette="Set3")
plt.title("Movie Duration by Type")
plt.xlabel("Type")
plt.ylabel("Duration (Minutes)")
plt.show()
```

 ### ðŸ” Insights
Boxplot of Movie Duration by Type:
The boxplot indicates that most movies fall within a duration range of 90-110 minutes. However, the presence of outliers suggests that there is a considerable variation in movie lengths, with a few movies being significantly longer or shorter than the average.

```python
#Correlation Heatmap
# Select numerical columns (Release_year and Minutes)
plt.figure(figsize=(8, 6))
numerical_df = df[["release_year", "Minutes"]].dropna()
corr_matrix = numerical_df.corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap between Release Year and Duration")
plt.show()
```

 ### ðŸ” Insights
 * Movies and TV shows released in different years have similar durations, suggesting that duration is not tied to the release year.
 * The length of movies (usually 90-120 minutes) and TV shows (usually 1-2 seasons) is driven by genre and format, not by the release year.

```python
#Genres by Country
#common genres in the US and India
# Filter rows for US and India
us_data = df[df['country'].str.contains('United States', na=False)]
india_data = df[df['country'].str.contains('India', na=False)]
us_genres = us_data['listed_in'].str.split(',').explode().str.strip().value_counts().head(10)
india_genres = india_data['listed_in'].str.split(',').explode().str.strip().value_counts().head(10)
plt.figure(figsize=(14, 6))

# US Genres
plt.subplot(1, 2, 1)
sns.barplot(y=us_genres.index, x=us_genres.values, palette="viridis")
plt.title("Top Genres in the United States")
plt.xlabel("Number of Titles")
plt.ylabel("Genres")

# India Genres
plt.subplot(1, 2, 2)
sns.barplot(y=india_genres.index, x=india_genres.values, palette="magma")
plt.title("Top Genres in India")
plt.xlabel("Number of Titles")
plt.ylabel("Genres")
plt.tight_layout()
plt.show()
```

 ### ðŸ” Insights
 * United States:
   * Dominant genres are Drama, Comedy, and Action.
   * There is significant content in Thrillers, Horror, and Romantic Movies, appealing to a broad audience.
* India:
    * Regional genres like Dramas, Thrillers, and Romantic Movies dominate.
    * Content often reflects local culture, traditions, and storytelling styles.

```python
# Group by Country and Type (Movies/TV Shows)
content_type_by_country = (
    df.groupby(['country', 'type'])
    .size()
    .unstack(fill_value=0)
    .loc[['United States', 'India']]  # Focus on US and India
)

# Plotting
content_type_by_country.plot(kind='bar', figsize=(10, 6), stacked=True, colormap="tab10")
plt.title("Distribution of Movies and TV Shows in the US and India")
plt.xlabel("Country")
plt.ylabel("Number of Titles")
plt.xticks(rotation=0)
plt.legend(title="Type")
plt.show()
```

 ### ðŸ” Insights 
 * In the United States, there is a balanced distribution of Movies and TV Shows.
 * In India, there is a higher focus on Movies, which aligns with the country's strong movie-watching culture (e.g., Bollywood).

```python
# Convert 'date_added' to datetime and extract the month
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['Month_Added'] = df['date_added'].dt.month
month_names = {1: 'January', 2: 'February', 3: 'March', 4: 'April', 
               5: 'May', 6: 'June', 7: 'July', 8: 'August', 
               9: 'September', 10: 'October', 11: 'November', 12: 'December'}
df['Month_Added'] = df['Month_Added'].map(month_names)
```

```python
# Count titles by month
monthly_counts = df["Month_Added"].value_counts().reindex(list(month_names.values()))

# Plot the count of releases by month
plt.figure(figsize=(12, 6))
sns.barplot(x=monthly_counts.index, y=monthly_counts.values, palette="coolwarm")
plt.title("Number of Titles Added to Netflix by Month")
plt.xlabel("Month")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)
plt.show()
```

 ### ðŸ” Insights 
 * The highest number of titles were added to Netflix in July and then December.
 * The number of titles added each month is relatively consistent throughout the year, with a few exceptions.
 * Fewer titles were added in May compared to the months before.

## Business Insights

1. Content Growth
   * Netflix has dramatically increased its content library over the years, with a noticeable focus on TV shows in more recent times.
   * This reflects a shift in strategy to produce more episodic content, likely due to its ability to engage viewers for longer durations compared to movies.
2. Regional Preferences
   * Content preferences vary by region. For example:
     * US: Strong focus on mainstream genres like Drama, Comedy, and Action.
     * India: Growing interest in regional dramas, thrillers, and movies with local cultural appeal.
   * This highlights the importance of tailoring content to meet the tastes of specific countries or regions.
3. Timing of Releases
   * Movies and TV shows released during holiday periods (e.g., December or summer vacations) tend to have higher engagement and viewership.
   * This indicates that timing plays a critical role in maximizing the success of new releases.
4. Genre Preferences
   * The most popular types of shows and movies on Netflix are Drama, Comedy, Action, and Documentaries. However, new types of content, like regional shows (based on local cultures) and documentaries, are becoming more popular.
   * This means Netflix has a chance to offer more variety and cater to different viewer interests by creating content in these growing categories.

## Recommendations

1. Focus on TV Shows
   * Since TV shows keep viewers engaged for longer durations due to multiple seasons and episodes, Netflix should prioritize producing more high-quality TV shows.
   * For regions where TV shows are more popular (e.g., the US and Europe), Netflix can focus on developing premium episodic content in genres like Drama, Thriller, or Comedy.

2. Regional Customization
   * Expand content production tailored to specific countries based on their preferences:
     * India: Invest in regional-language dramas and thrillers.
     * South Korea and Japan: Produce more K-dramas and anime-style content to cater to their growing global popularity.
     * Brazil and Spain: Focus on Spanish-language thrillers and crime dramas, which have strong local and global appeal.
   * Collaborating with regional directors and writers can ensure authenticity and attract a larger audience.

3. Optimal Timing for Releases
    * Major content (e.g., blockbuster TV shows or movies) should be launched during holiday periods or peak seasons, such as:
       * December (winter holidays).
       * Summer vacations.
       * Special cultural or regional festivals (e.g., Diwali in India, Lunar New Year in Asia).
    * This will maximize initial viewership and social media buzz around the content.

4. Genre Diversification
    * While Drama and Comedy remain dominant, Netflix should explore underrepresented but growing genres such as:
       * Documentaries and biopics.
       * Fantasy and Sci-Fi for younger audiences.
       * Regional historical dramas or folklore-based series.
    * Investing in less-explored genres can help capture untapped audience segments and improve overall subscriber engagement.

5. Predictive Analysis for Content Success:
    * Utilize predictive analytics to evaluate the potential success of new content before production. This can include:
       * Assessing viewer interest in specific genres or themes based on past performance data.
       * Analyzing trends to predict demand for upcoming regional or global events (e.g., Olympics, elections, or cultural festivals).
    * This approach will help Netflix optimize its investment in content creation and align production with audience expectations.
